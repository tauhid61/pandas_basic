# Group By

    Example : Calculate avg population state wise using group by 

    Method 1 : Using groupby with user-defined function

```python
import pandas as pd
import numpy as np
df = df[df['SUMLEV'] == 50]
df_grouped = df.groupby('STNAME')
df_grouped.head(1)
for index, values in df_grouped:
    Avg = np.mean(values['CENSUS2010POP'])
    print( index , 'has : ', str(Avg) , 'population')
    
```

    Method 2 : using agg() function
    
    Note : Syntex for agg() is => agg({'Column name' : Method}) (for old version)
                               => agg([method1, method2, ...])  (for new version)
    

```python
df = pd.read_csv('census.csv')
df.groupby('STNAME').agg({'CENSUS2010POP' : np.average})
```

    Looking at our backpacking equipment DataFrame, suppose we are interested in finding our total weight for each category. Use groupby to group the dataframe, and apply a function to calculate the total weight (Weight x Quantity) by category.

```python
import pandas as pd
df_backpack = pd.read_csv('backpack.csv')
df_backpack.head(1)
```

```python
df_backpack.head(1)
```

```python
def total_weight(df_backpack):
    return sum(df_backpack['Quantity'] * df_backpack['Weight (oz.)'])
df_new = df_backpack.groupby('Category').apply(total_weight)
print(df_new)
```

```python
df_backpack.head(1)
```

    Groupby using multiple Columns - Single Index(Series groupby)

```python
df_backpack.set_index('Item').groupby(level = 0)['Category','Quantity', 'Weight (oz.)'].head(1)
```

        Groupby (multiple Columns) - Multiple Index (Dataframe groupby)

```python
(df_backpack.set_index(['Item','Category'])
 .groupby(level = 0)['Quantity', 'Weight (oz.)']
 .agg([np.sum, np.mean])
).head(1)
```

    Example 3 : Reading data from 'census.csv' file, excluding 'SUMLEV'==50, set index to 'STNAME' and Groupby using 'CENSUS2010POP' and calculate their avrage. and then Categorize the data in 10 Categories.

```python
import numpy as np
df = pd.read_csv('census.csv')
df = df[df['SUMLEV'] == 50]
new_df = (df.set_index('STNAME')
 .groupby(level = 0)['CENSUS2010POP']
 .agg([np.mean])
)
pd.cut(new_df['mean'], 10 , labels = ['Category 1','Category 2','Category 3','Category 4','Category 5'
                                      ,'Category 6','Category 7','Category 8','Category 9','Category 10'])
```

##### Example:
    Suppose we want to simulate the probability of flipping a fair coin 20 times, and getting a number greater than or equal to 15. Use np.random.binomial(n, p, size) to do 10000 simulations of flipping a fair coin 20 times, then see what proportion of the simulations are 15 or greater.

    Syntex:
    numpy.random.binomial(n, p, size=None)
    n trials and p probability of success where n an integer >= 0 and p is in the interval [0,1]

```python
import numpy as np
x = np.random.binomial(20,0.5, 10000)
(x>=15).mean()
```

    chances of back to back tornado happening in two days in 1000000 years

```python
probability  = 0.01
tornado_event = np.random.binomial(1, probability, 1000000)
tornado_twoday= 0

for i in range(1,len(tornado_event)-1):
    if tornado_event[i] == 1 and tornado_event[i-1] == 1:
        tornado_twoday += 1
        
print('{} numbers of tornado will happen two days in a row in about {} years'.format(tornado_twoday,1000000/365))
```
